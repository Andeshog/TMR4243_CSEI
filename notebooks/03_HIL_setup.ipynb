# Connection setup

Current setup includes a Raspberry Pi with Ubuntu 20.04 installation. The 
default user name and the password for the raspberry pi is given in the table 
below.

|User Name|Password|
|---------|--------|
|`pi`     |`raspberry`|

## SSH connection to the single board computer

There are several ways to get connected to the single board computer. The first
and the basic way is to do it from terminal of the Ubuntu Laptop. So, open op
a new terminal with keystroke `CTRL`+`ALT`+`T` and enter the following command:

> Note the the following command assumes the network configuration described 
below.

```bash
ssh pi@10.42.0.50
```
This command will ask you for the password, enter the password above.


## Remote connection with Visual Studio Code

# Network setup
The computations will be done on the Single Board Computer. In order for us to 
grab the ROS topic streams out of single board computer, we have to do some 
custom network setup.

_For further information, please consult to [ROS WiKi - Network Setup](http://wiki.ros.org/ROS/NetworkSetup)._


## Network setup for Hardware-In-the-Loop


### Connection between Laptop and Single Board Computer 

The connection between the Single Board Computer and the laptop established 
through an ethernet cable. Once plugged in, the laptop takes the IP address 
`10.42.0.1` and the Single Board Computer takes the IP address `10.42.0.50`.


|Device                | IP Address     |
|----------------------|----------------|
|Ubuntu Laptop         | `10.42.0.1`    |
|Single Board Computer | `10.42.0.50`   |

> For fallback, Single Board Computer is configured to listed for DHCP servers 
as well. When plugged in to another router or network via cable it can get other 
IP addresses as well.

### Setting environment variables for ROS multi-machine connection

To exchange ROS messages between multiple machines, we have to notify ROS nodes
about IP addresses. We do it by setting several environment variables.

These environment variables are `ROS_MASTER_URI` and `ROS_IP` address.

|Device| Command|
|------|--------|
|Ubuntu Laptop|<code>export ROS_IP=10.42.0.1<br>export ROS_MASTER_URI=http://10.42.0.50:11311</code>|
|Single Board Computer|<code>export ROS_IP=10.42.0.50</code>|

You can double check to see if these variables are set by typing the following 
commands

```bash
printenv | grep -i ros_
```

### Pro Tip

In this lab, we are using `bash` as our [shell](https://linuxcommand.org/lc3_lts0010.php).
Every time we open up a new terminal several files are being executed. The last
file to executed is the `~/.bashrc` file. So, if we add respective lines to the
`~/.bashrc` files both on the Ubuntu Laptop and the Single Board Computer, we
won't need to enter the commands that begin with `export`.

> In linux, tilda (`~`) represents the home directory.
